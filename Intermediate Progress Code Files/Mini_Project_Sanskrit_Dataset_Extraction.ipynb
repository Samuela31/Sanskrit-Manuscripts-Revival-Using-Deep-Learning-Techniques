```python
import re
import string
```

```python
# Text preprocessing functions
def clean_and_tokenize_text(text):
    # Remove unwanted characters, punctuation, and whitespace.
    text = re.sub(f"[{string.punctuation}]", "", text)
    text = " ".join(text.split())
    return text

# Specify the folder path where your image and text files are located.
folder_path = '/content/OCR-Images-Annotation/BHSv1S1'

# Process text files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        # Text preprocessing
        text_filename = os.path.splitext(filename)[0] + '.txt'
        text_path = os.path.join(folder_path, text_filename)

        with open(text_path, 'r', encoding='utf-8') as text_file:
            text_data = text_file.read()

        cleaned_and_tokenized_text = clean_and_tokenize_text(text_data)

        # Save the preprocessed text
        output_tokenized_text_path = os.path.join(folder_path, 'cleaned_tokenized_' + text_filename)

        with open(output_tokenized_text_path, 'w', encoding='utf-8') as output_file:
            output_file.write(cleaned_and_tokenized_text)  # Save the clean and tokenized text from text files

print("Data preprocessing completed for files in BHSv1S1.")
```
