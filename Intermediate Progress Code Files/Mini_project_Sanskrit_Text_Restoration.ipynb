```python
from transformers import AutoTokenizer, AutoModelForMaskedLM
from torch.utils.data import DataLoader, TensorDataset
import torch
```

```python
# Folder path containing Sanskrit text files
folder_path = '/content/OCR-Images-Annotation/GG1'

# Function to read and tokenize text files
def read_and_tokenize_files(folder_path, tokenizer, max_length=512):
    input_ids_list, attention_mask_list, token_type_ids_list = [], [], []
    lim_files=0
    for file_name in os.listdir(folder_path):
        if file_name.startswith("preprocessed_tokenized_") and file_name.endswith(".txt") and lim_files<10:
            file_path = os.path.join(folder_path, file_name)
            lim_files= lim_files+1 #limiting number of files used for training
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()

                # Tokenize the text using a sliding window
                tokens = tokenizer(content, return_tensors="pt", padding=True, truncation=True, max_length=max_length)

                input_ids_list.append(tokens["input_ids"].squeeze(0))  # Remove the added batch dimension
                attention_mask_list.append(tokens["attention_mask"].squeeze(0))
                token_type_ids_list.append(tokens["token_type_ids"].squeeze(0))  # Add token type IDs

    # Pad sequences to a consistent length
    input_ids_tensor = torch.nn.utils.rnn.pad_sequence(input_ids_list, batch_first=True)
    attention_mask_tensor = torch.nn.utils.rnn.pad_sequence(attention_mask_list, batch_first=True)
    token_type_ids_tensor = torch.nn.utils.rnn.pad_sequence(token_type_ids_list, batch_first=True)

    return TensorDataset(input_ids_tensor, attention_mask_tensor, token_type_ids_tensor)

# Load pre-trained Sanskrit ALBERT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("surajp/albert-base-sanskrit")
model = AutoModelForMaskedLM.from_pretrained("surajp/albert-base-sanskrit")

# Read and tokenize files for training
dataset = read_and_tokenize_files(folder_path, tokenizer)

# Create DataLoader for training
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# Fine-tune the model
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
for epoch in range(3):  # Adjust the number of epochs
    for batch in dataloader:
        optimizer.zero_grad()
        input_ids, attention_mask, token_type_ids = batch
        outputs = model(input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids, labels=input_ids)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
```

```python
# Sanskrit text input with a masked position
sanskrit_text = "तत्र सत्यस्य परमं निधानं [MASK] यः न प्रियः ।"

# Tokenize the input
tokenized_input = tokenizer(sanskrit_text, return_tensors="pt", padding=True, truncation=True)

# Forward pass through the model
with torch.no_grad():
    outputs = model(**tokenized_input)

# Get the predicted logits for the masked position
logits = outputs.logits
masked_index = tokenized_input["input_ids"].view(-1).tolist().index(tokenizer.mask_token_id)
predicted_logits = logits[0, masked_index, :]

# Get the predicted word
predicted_token_id = torch.argmax(predicted_logits).item()
predicted_word = tokenizer.decode(predicted_token_id)

print("Predicted Word:", predicted_word)
```

```python
# Sanskrit text input with a masked position
sanskrit_text = "तत्र सत्यस्य परमं निधानं [MASK] यः न प्रियः ।"

# Tokenize the input
tokenized_input = tokenizer(sanskrit_text, return_tensors="pt", padding=True, truncation=True)

# Forward pass through the model
with torch.no_grad():
    outputs = model(**tokenized_input)

# Get the predicted logits for the masked position
logits = outputs.logits
masked_index = tokenized_input["input_ids"].view(-1).tolist().index(tokenizer.mask_token_id)
predicted_logits = logits[0, masked_index, :]

# Apply softmax to get probabilities
predicted_probs = torch.nn.functional.softmax(predicted_logits, dim=-1)

# Get the top 3 predicted words with probabilities
top_k = 3
top_predictions = torch.topk(predicted_probs, k=top_k)
predicted_words = [tokenizer.decode(token_id.item()) for token_id in top_predictions.indices]
prediction_probs = top_predictions.values.tolist()

# Display the top 3 predicted words with probabilities
for i, (word, prob) in enumerate(zip(predicted_words, prediction_probs), 1):
    print(f"Top {i} Predicted Word: {word}, Probability: {prob}")
```
